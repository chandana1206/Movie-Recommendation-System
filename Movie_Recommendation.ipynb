```python
movies_df = movies_df[['Movie_Title', 'Year', 'Genres', 'Actors', 'Director','Rating']]
```

```python
# Combine the genres, actors, and director into a single column
movies_df['tags'] = movies_df['Genres'] + ' ' + movies_df['Actors'] + ' ' + movies_df['Director']
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
import string
def lemmatization(tags):
    words = []
    for word in tags.split():
        # Lemmatize the word and remove punctuation
        word = word.strip(string.punctuation)
        words.append(lemmatizer.lemmatize(word))
    return ' '.join(words)

movies_df['tags'] = movies_df['tags'].apply(lemmatization)
```

```python
# Term Frequency-Inverse Document Frequency 
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Initialize the TF-IDF Vectorizer
tfidf = TfidfVectorizer(stop_words='english')

# Transform the 'tags' column into the TF-IDF matrix
tfidf_matrix = tfidf.fit_transform(movies_df['tags'])

tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate the cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

cosine_sim
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    title = title.strip().lower()

    # Get the index of the movie that matches the lowercase input title
    idx = movies_df.index[movies_df['Movie_Title'].str.strip().str.lower() == title].tolist()

    # Check if the title exists
    if not idx:
        return "Movie not found!"

    idx = idx[0]  # Get the index from the list

    # Get pairwise similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort movies by similarity score (highest to lowest)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the indices of the top 10 most similar movies
    top_indices = [i[0] for i in sim_scores[1:6]]  # Exclude the first one (the movie itself)

    # Get the top 10 movie titles and their corresponding years
    recommendations = movies_df[['Movie_Title', 'Year']].iloc[top_indices]

    recommendations.reset_index(drop=True, inplace=True)
    recommendations.index += 1

    return recommendations
```

```python
get_recommendations(" annabelle comes home")
```

```python
def get_top_movies_with(genre=None, actor=None, director=None, num_recs=10):
    # Start with all movies
    filtered_movies = movies_df

    # Filter by genre if provided
    if genre:
        filtered_movies = filtered_movies[filtered_movies['Genres'].str.contains(genre.lower(), case=False, na=False)]

    # Filter by actor if provided
    if actor:
        filtered_movies = filtered_movies[filtered_movies['Actors'].str.contains(actor.lower(), case=False, na=False)]

    # Filter by director if provided
    if director:
        filtered_movies = filtered_movies[filtered_movies['Director'].str.contains(director.lower(), case=False, na=False)]

    # If no movies are found after all filters, return a message
    if filtered_movies.shape[0] == 0:
        return "No Movies Found"

    # Sort by ratings in descending order
    filtered_movies = filtered_movies.sort_values(by='Rating', ascending=False)

    # Return the top N movie titles
    recommendations=filtered_movies[['Movie_Title', 'Year']].head(num_recs)
    recommendations.reset_index(drop=True, inplace=True)
    recommendations.index += 1

    return recommendations
```

```python
get_top_movies_with(genre="horror")
```
